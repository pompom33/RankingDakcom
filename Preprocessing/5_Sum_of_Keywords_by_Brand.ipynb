# 브랜드 별 키워드 개수 합산

```python
import pandas as pd
```

```python
columns = [
    'flavor_positive', 
    'flavor_negative', 
    'texture_positive', 
    'texture_negative', 
    'ingredient', 
    'diet', 
    'exercise', 
    'reorder', 
    'convenience', 
    'delivery_positive',
    'delivery_negative'
]
```

```python
files = [
    "[랭커]키워드요약_수정",
    "[러브잇]키워드요약",
    "[맛있닭]키워드_요약",
    "[아임닭]키워드요약",
    "[잇메이트]키워드_요약",
    "[잇메이트]키워드_요약",
    "[함께하닭]키워드요약",
    "[햇살닭]키워드요약"
]
```

```python
total_df = pd.DataFrame()
```

```python
for file in files:
    file_path = f"./Keywords/{file}.csv"
    df = pd.read_csv(file_path, encoding="utf-8")
    
    for column in columns:
        total_dict = {} 
        
        for i in range(len(df)):
            data_str = df[column].iloc[i]
            data_dict = eval(data_str)
            
            for key, value in data_dict.items():
                if key in total_dict:
                    total_dict[key] += value
                else:
                    total_dict[key] = value
        
        total_data = str(total_dict)
        total_df.loc[file, column] = total_data
```

```python
total_df.to_csv("./Keywords/total_keyword_랭커수정.csv", encoding="utf-8")
```
