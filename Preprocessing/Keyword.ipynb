## 브랜드별 키워드

```python
import pandas as pd
import os
from collections import defaultdict

# 키워드 카운팅을 딕셔너리 형태로
def count_keywords_in_text(text, keyword_categories):
    category_counts = {category: defaultdict(int) for category in keyword_categories}
    
    for category, keywords in keyword_categories.items():
        for keyword in keywords:
            category_counts[category][keyword] += text.lower().count(keyword.lower())
    
    return category_counts

brand_keyword_counts = []

for brand, files in brands.items():
    #토탈 키워드 카운트 설정.
    total_keyword_count = {category: defaultdict(int) for category in keyword_categories}
    
    for file in files:
        file_path = f'./data/3_번역_완료/{brand}_번역/{file}_translated.csv'
        
        if os.path.exists(file_path):
            # 파일 읽기
            df = pd.read_csv(file_path, encoding='utf-8')
            
            # '맛있닭'과 '잇메이트' 브랜드의 경우 'translation' 리뷰 컬럼
            if brand in ['맛있닭', '잇메이트']:
                review_column = 'translation'
            else:
                review_column = 'translated_review'
            
            # 리뷰가 들어있는 컬럼 순회
            for text in df[review_column]:
                category_counts = count_keywords_in_text(text, keyword_categories)
                for category, counts in category_counts.items():
                    for keyword, count in counts.items():
                        total_keyword_count[category][keyword] += count
        else:
            print(f"파일을 찾을 수 없습니다: {file_path}")
    
    # 브랜드별로 저장.
    brand_data = {'brand': brand}
    for category, counts in total_keyword_count.items():
        brand_data[category] = dict(counts)
    
    brand_keyword_counts.append(brand_data)

# 데이터프레임 설정.
results_df = pd.DataFrame(brand_keyword_counts)

# CSV 파일로 저장.
results_df.to_csv('./data/3_번역_완료/브랜드별_키워드.csv', index=False)
```

## 재구매율 계산하기

```python
import pandas as pd

results = []

for brand, products in brands.items():
    # 변수 설정
    total_reviews_sum = 0
    repeated_purchases_sum = 0
    
    for product in products: 
        df = pd.read_csv(f'./data/3_번역_완료/{brand}_번역/{product}_translated.csv', encoding='utf-8')
        
        # 고유 ID값을 총 리뷰 수로 계산
        total_reviews = df['id'].nunique()
        
        # 재구매 계산. ID 카운팅이 2이상인 것들
        id_counts = df['id'].value_counts()
        repeated_purchases = id_counts[id_counts > 1].count()
        
        # 제품별 합계 누적시켜서 브랜드 총 분모값 분자값 카운팅
        total_reviews_sum += total_reviews
        repeated_purchases_sum += repeated_purchases

    # 브랜드의 재구매율 계산하기(백분율)
    repeat_purchase_rate_sum = (repeated_purchases_sum / total_reviews_sum) * 100 if total_reviews_sum > 0 else 0
    
    # 결과 리스트에 가가
    results.append([brand, repeated_purchases_sum, repeat_purchase_rate_sum])

# 데이터프레임으로
summary_df = pd.DataFrame(results, columns=['brand', '재구매_수', '재구매율'])
summary_df.to_csv('./data/재구매율.csv',index=False)
```

## 번역한 리뷰 통합

```python
import pandas as pd
import os

# 결과를 저장할 리스트를 초기화
combined_data = []

# 브랜드별로 파일 순회
for brand, files in brands.items():
    for file in files:
        # 파일 경로 설정
        file_path = f'./data/3_번역_완료/{brand}_번역/{file}_translated.csv'

        if os.path.exists(file_path):
            # 파일 읽기
            df = pd.read_csv(file_path, encoding='utf-8')

            # '맛있닭'과 '잇메이트' 브랜드의 경우 'translation' 열을 사용
            if brand in ['맛있닭', '잇메이트']:
                review_column = 'translation'
            else:
                review_column = 'translated_review'

            # 통합된 리뷰 열 만들기
            if 'translation' in df.columns and 'translated_review' in df.columns:
                df['translated_review'] = df['translated_review'].combine_first(df['translation'])
            elif 'translation' in df.columns:
                df['translated_review'] = df['translation']

            # 필요한 열만 데이터프레임에 추가
            if 'translated_review' in df.columns:
                df_filtered = df[['translated_review']].copy()
                df_filtered['brand'] = brand
                df_filtered['file'] = file


                # 결과 리스트에 추가
                combined_data.append(df_filtered)
        else:
            print(f"파일을 찾을 수 없습니다: {file_path}")

# 리스트를 데이터프레임으로 변환
combined_df = pd.concat(combined_data, ignore_index=True)

# 브랜드와 파일 컬럼을 앞으로, 번역 리뷰를 뒤로
columns = ['brand', 'file',] + [col for col in combined_df.columns if col not in ['brand','file']]
combined_df = combined_df[columns]

# 데이터프레임을 CSV 파일로 저장
combined_df.to_csv('./data/combined_reviews.csv', index=False)
```
