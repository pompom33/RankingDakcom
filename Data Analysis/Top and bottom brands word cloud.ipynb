```python
#잇메이트 1위
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# CSV 파일 로드
file_path = './total_word_count/[잇메이트]키워드_요약.csv'
df = pd.read_csv(file_path)

# 모든 키워드를 결합한 텍스트 생성
word_count = {}

for column in df.columns[1:]:  # 'title' 열을 제외한 나머지 열
    for row in df[column].dropna():
        keywords_dict = eval(row)
        for keyword, count in keywords_dict.items():
            if keyword in word_count:
                word_count[keyword] += int(count)
            else:
                word_count[keyword] = int(count)

# 단어를 등장 횟수에 따라 내림차순으로 정렬
sorted_word_count = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

# 정렬된 결과 출력
for word, count in sorted_word_count:
    print(f"'{word}': {count}번 등장")

# 워드 클라우드 생성
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color='white', 
    colormap='Oranges',  
    contour_color='black',  # 윤곽선 색상
    contour_width=1  # 윤곽선 두께
).generate_from_frequencies(dict(sorted_word_count))

# 생성된 워드 클라우드를 표시
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축을 숨깁니다.
plt.show()
```
