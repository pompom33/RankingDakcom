2.상위 브랜드, 하위 브랜드 그룹 내 긍정 부정 키워드 비교

```python
import matplotlib.pyplot as plt
import pandas as pd

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows에서는 'Malgun Gothic' 폰트를 사용
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호가 깨지는 문제 방지


# 긍정키워드, 부정키워드
positive_keywords = ['flavor_positive', 'texture_positive', 'ingredient', 'diet', 'exercise', 'reorder', 'convenience', 'delivery_positive','cost_positive','volume_positive']
negative_keywords = ['flavor_negative', 'texture_negative', 'delivery_negative','volume_negative']


# 브랜드를 상위권(Top-tier)과 하위권(Bottom-tier)으로 분류
top_tier_brands = ['잇메이트', '맛있닭', '한끼통살','랭커']
bottom_tier_brands = ['러브잇', '햇살닭', '아임닭', '함께하닭']

# 선택된 브랜드에 대해 데이터를 필터링
top_tier_data = data[data['brand'].isin(top_tier_brands)]
bottom_tier_data = data[data['brand'].isin(bottom_tier_brands)]

# 전체 데이터 포인트 수 계산
total_top_tier = len(top_tier_data)
total_bottom_tier = len(bottom_tier_data)




# 그룹 내 키워드 빈도수를 합한 값으로 나누어, 각 키워드가 그룹 내에서 차지하는 비율을 계산
top_tier_positive_summary = top_tier_data[positive_keywords].sum(axis=0) / top_tier_data[positive_keywords].sum(axis=0).sum()
bottom_tier_positive_summary = bottom_tier_data[positive_keywords].sum(axis=0) / bottom_tier_data[positive_keywords].sum(axis=0).sum()

top_tier_negative_summary = top_tier_data[negative_keywords].sum(axis=0) / top_tier_data[negative_keywords].sum(axis=0).sum()
bottom_tier_negative_summary = bottom_tier_data[negative_keywords].sum(axis=0) / bottom_tier_data[negative_keywords].sum(axis=0).sum()

# 시각화를 위해 데이터프레임으로 결합
positive_summary_df = pd.DataFrame({
    '상위권 브랜드': top_tier_positive_summary,
    '하위권 브랜드': bottom_tier_positive_summary
})

negative_summary_df = pd.DataFrame({
    '상위권 브랜드': top_tier_negative_summary,
    '하위권 브랜드': bottom_tier_negative_summary
})

# 긍정적 키워드를 시각화
plt.figure(figsize=(10, 7))
ax = positive_summary_df.plot(kind='bar', width=0.8)
plt.title('상위권 브랜드 vs 하위권 브랜드 (긍정적 키워드)', fontsize=14)
plt.ylabel('상대 빈도수 (그룹 내 비율)', fontsize=13)
plt.xlabel('키워드', fontsize=13)
plt.legend(title='그룹', loc='upper right')
plt.tight_layout()
plt.xticks(size=10, rotation=45, ha="center")

# 막대 위에 숫자 표기
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# 긍정적 키워드에 대한 그래프 표시
plt.show()

# 부정적 키워드를 시각화
plt.figure(figsize=(10, 7))
ax = negative_summary_df.plot(kind='bar', width=0.8)
plt.title('상위권 브랜드 vs 하위권 브랜드 (부정적 키워드)', fontsize=14)
plt.ylabel('상대 빈도수 (그룹 내 비율)', fontsize=13)
plt.xlabel('키워드', fontsize=13)
plt.legend(title='그룹', loc='upper left')
plt.tight_layout()
plt.xticks(size=10, rotation=45, ha="center")

# 막대 위에 숫자 표기
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# 부정적 키워드에 대한 그래프 표시
plt.show()
```
